```python
import pandas as pd
```

```python
df = pd.read_csv('/content/drive/My Drive/ops_segmentation-main/current_stock.csv')
```

```python
df.info()
```

```python
# Assuming 'df' is your original DataFrame with the 'outlet' column intact

# Filter the DataFrame for rows where strItemBarcode is '0' or NULL
df_filtered = df[(df['strItemBarcode'].isnull()) | (df['strItemBarcode'] == '0')]

# Drop duplicates based on 'outlet' and 'Item_code' to get unique combinations
df_unique_filtered = df_filtered.drop_duplicates(subset=['intWarehouseId', 'intItemId'])

# Display the result
print(df_unique_filtered.info())
print(df_unique_filtered)
```

```python
df_unique_filtered.head(5)
```

```python
df2 = pd.read_csv('/content/drive/My Drive/ops_segmentation-main/outlets.csv')
```

```python
df2.head(5)
```

```python
df_unique_filtered.rename(columns={'intWarehouseId': 'outlet'}, inplace=True)
```

```python
# Assuming df2 contains the columns 'outlet' and 'Outlet_name'

# Merge df_unique_filtered with df2 based on the 'outlet' column
df_result = pd.merge(df_unique_filtered, df2[['outlet', 'outlet_name']], on='outlet', how='left')

# Display the resulting DataFrame
print(df_result.info())
df_result.head(5)
```

```python
df_result.to_excel('df_result.xlsx')  # Default is index=True
print("Data exported to df_result.xlsx successfully!")
```

```python
# Assume df_result is the DataFrame we previously created

# Sort df_result if necessary (optional, depending on how you want to pick the "first" instance)
# For example, sorting by 'outlet' to preferentially select from lower numbered outlets first
df_result_sorted = df_result.sort_values(by='outlet')

# Drop duplicates based on 'Item_code', keeping the first occurrence
df_final = df_result_sorted.drop_duplicates(subset='intItemId', keep='first')

# Display the final DataFrame to be sent to outlet managers
print(df_final.info())
df_final.head(5)
```

```python
df_final.to_excel('df_final.xlsx')  # Default is index=True
print("Data exported to df_final.xlsx successfully!")
```

```python
# List of outlet names to exclude
excluded_outlets = [
    "Apon Wellbeing Limited Corporate",
    "Odyssey Crafts (Pvt.) Ltd.",
    "SQ Station",
    "Ananta Denim Technology Limited",
    "SQ Celsius Limited",
    "Ananta Casual Wear Ltd.",
    "Elite Garments Ltd"
]

# Filter df_result to exclude these outlet names
df_filtered = df_result[~df_result['outlet_name'].isin(excluded_outlets)]

# For each item_code in df_final that needs to be replaced, find an alternative
replacement_needed = df_final[df_final['outlet_name'].isin(excluded_outlets)]

# Find replacements
replacements = pd.DataFrame()

for index, row in replacement_needed.iterrows():
    alternative = df_filtered[(df_filtered['intItemId'] == row['intItemId']) & (df_filtered['strItemBarcode'].isin(['0', None]))].head(1)
    replacements = pd.concat([replacements, alternative], ignore_index=True)

# Drop the rows needing replacement from df_final
df_final = df_final[~df_final['intItemId'].isin(replacement_needed['intItemId'])]

# Concatenate the replacements into df_final
df_final = pd.concat([df_final, replacements], ignore_index=True)

# Removing any potential duplicates that might have been introduced
df_final = df_final.drop_duplicates(subset=['intItemId'], keep='first')

# Display the final DataFrame
print(df_final.info())
df_final.head(5)
```

```python
df_final.to_excel('df_final3.xlsx')  # Default is index=True
print("Data exported to df_final2.xlsx successfully!")
```
