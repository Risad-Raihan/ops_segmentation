```python
import pandas as pd

# Assuming df_Lida and df_prod are already loaded DataFrames

# Merging df_Lida with df_prod on the condition that Item Name matches Product_Name
result_df = pd.merge(df_Lida, df_prod, left_on='Item Name', right_on='Product_Name', how='inner')

# Selecting the required columns including the ones from df_prod
result_df = result_df[['Date', 'Outlet', 'Item Code', 'Item Name', 'Item Sub Category', 'Big Block', 'Total Quantity', 'Total Sales Amount', 'Total_COGS', 'Final_COGS', 'Manufacturer_Name', 'Sub_Category']]

# Display the resulting DataFrame
print(result_df.head())
```

```python
result_df.info()
```

```python
import pandas as pd

# Assuming result_df is already defined as per previous operations

# Calculate Margin(%) for each row
result_df['Margin(%)'] = ((result_df['Total Sales Amount'] - result_df['Final_COGS']) / result_df['Total Sales Amount']) * 100

# Display the updated DataFrame with the new Margin(%) column
print(result_df.head())
```

```python
# Save DataFrame to Excel file in Colab's local environment
result_df.to_excel('/content/Lida_margin.xlsx', index=False)
```

```python
import pandas as pd

# Assuming result_df is already defined and may have a Margin(%) column

# Check if 'Margin(%)' column exists and drop it if present
if 'Margin(%)' in result_df.columns:
    result_df = result_df.drop(columns=['Margin(%)'])

# Display the DataFrame after dropping the Margin(%) column
print(result_df.head())
```

```python
import pandas as pd

# Assuming result_df is already defined

# Convert Date column to datetime if not already
result_df['Date'] = pd.to_datetime(result_df['Date'])

# Extract month from the Date column
result_df['Month'] = result_df['Date'].dt.month

# Dropping the original Date column as it is no longer needed
result_df = result_df.drop(columns=['Date'])

# Grouping by the Month and other categorical columns, you can adjust the aggregation as needed
# For numerical columns like Total Quantity, Total Sales Amount, Final_COGS, you might want to sum these
grouped_df = result_df.groupby(['Month', 'Outlet', 'Item Code', 'Item Name' ]).agg({
    'Total Quantity': 'sum',
    'Total Sales Amount': 'sum',
    'Total_COGS': 'sum',
    'Final_COGS': 'sum'       # Assuming you want to just keep the first occurrence
}).reset_index()

# Display the grouped DataFrame
print(grouped_df.head())
```

```python
grouped_df.info(0)
```

```python
import pandas as pd

# Assuming grouped_df is already defined as per previous operations

# Calculate Margin(%) for each row in the grouped DataFrame
grouped_df['Margin(%)'] = ((grouped_df['Total Sales Amount'] - grouped_df['Final_COGS']) / grouped_df['Total Sales Amount']) * 100

# Display the updated DataFrame with the new Margin(%) column
print(grouped_df.head())
```

```python
# Save DataFrame to Excel file in Colab's local environment
grouped_df.to_excel('/content/Lida_margin2.xlsx', index=False)
```
