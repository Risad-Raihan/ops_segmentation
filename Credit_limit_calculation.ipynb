```python
import pandas as pd
import numpy as np

# Assuming merged_df is your dataframe

# Update the Monthly_Salary where it's 0 or NaN
merged_df['Monthly_Salary'] = np.where(
    (merged_df['Monthly_Salary'].isna()) | (merged_df['Monthly_Salary'] == 0),
    merged_df['Credit_Limit'] * 3,
    merged_df['Monthly_Salary']
)
```

```python
import pandas as pd
from datetime import datetime

# Assuming merged_df is your dataframe and Joining_Date is in string format, first convert it to datetime if needed
merged_df['Joining_Date'] = pd.to_datetime(merged_df['Joining_Date'], errors='coerce')

# Get today's date
today = pd.Timestamp(datetime.today())

# Calculate the difference in months and create a new column called 'Tenure'
merged_df['Tenure'] = (today.year - merged_df['Joining_Date'].dt.year) * 12 + (today.month - merged_df['Joining_Date'].dt.month)

# Fill NaN values with 0 (in case of missing Joining_Date values)
merged_df['Tenure'] = merged_df['Tenure'].fillna(0).astype(int)
```
